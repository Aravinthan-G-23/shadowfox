### *Data Visualization Using plotly*

## *Line PLot*

```python
import plotly.express as px #plotly library for data visualization
import pandas as pd

df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Sales": [100, 150, 130, 180]
})

fig = px.line(df, x="Month", y="Sales",
              title="Monthly Sales Trend") #line plot
fig.show()
```

## *Scatter Plot*

```python
import plotly.express as px

df = px.data.iris() #load iris dataset

fig = px.scatter(df,x="sepal_width",y="sepal_length",color="species")  #Scatter Plot

fig.show()
```

### *Bar Chart*

```python
import plotly.express as px

df = px.data.tips() #tips dataset

fig = px.bar(df,x="day",y="total_bill",color="total_bill") #Bar Chart

fig.show()
```

## *Histogram*

```python
import plotly.express as px

df = px.data.tips() #tips Dataset

fig = px.histogram(df,x="total_bill",nbins=20) #Histogram

fig.show()
```

## *Pie Chart*

```python
import plotly.express as px

data = {"Category": ["A", "B", "C"],"Values": [30, 40, 30]}

fig = px.pie(names=data["Category"],values=data["Values"]) #Pie Chart

fig.show()
```

## *Bokeh*

## *Line Plot*

```python
from bokeh.plotting import figure, show #Bokeh Library

x = [1, 2, 3, 4, 5]
y = [2, 5, 8, 2, 7]

p = figure(title="Line Plot") 

p.line(x, y, line_width=2) #Line Plot

show(p)
```

## *Scatter Plot*

```python
from bokeh.plotting import figure, show

x = [1, 2, 3, 4, 5]
y = [3, 6, 2, 8, 4]

p = figure(title="Scatter Plot")

p.scatter(x, y, size=10) #Scatter Plot

show(p)
```

## *Bar Chart*

```python
from bokeh.plotting import figure, show

fruits = ["Apple", "Banana", "Orange"]
counts = [20, 35, 15]

p = figure(x_range=fruits,title="Fruit Sales")

p.vbar(x=fruits,top=counts,width=0.5) #Vertical Bar Chart

show(p)
```

## *Histogram*

```python
import numpy as np
from bokeh.plotting import figure, show

data = np.random.randn(1000)

hist, edges = np.histogram(data,bins=20)

p = figure(title="Histogram")

p.quad(top=hist,bottom=0,left=edges[:-1],right=edges[1:]) #Histogram

show(p)
```

## *Pie Chart*

```python
from math import pi
from bokeh.plotting import figure, show

categories = ["A", "B", "C"]
values = [30, 40, 30]

# Different colors for each slice
colors = ["red", "blue", "green"]

angles = [v / sum(values) * 2 * pi for v in values]

p = figure(title="Pie Chart",width=500,height=500)

start = 0

for angle, color, category in zip(angles, colors, categories):
    end = start + angle

    p.wedge(x=0,y=1,radius=0.4,start_angle=start,end_angle=end,
        fill_color=color,
        line_color="white",
        legend_label=category
    )

    start = end

p.legend.location = "top_right"

show(p)
```
